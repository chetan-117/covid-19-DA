Q. 1) Show the number of Confirmed, Deaths and Recovered cases in each Region.

Q. 2) Remove all the records where the Confirmed Cases is Less Than 10.

Q. 3) In which Region, maximum number of Confirmed cases were recorded ?

Q. 4) In which Region, minimum number of Deaths cases were recorded ?

Q. 5) How many Confirmed, Deaths & Recovered cases were reported from India till 29 April 2020 ?

Q. 6-A ) Sort the entire data wrt No. of Confirmed cases in ascending order.

Q. 6-B ) Sort the entire data wrt No. of Recovered cases in descending order.


```python
import pandas as pd

data = pd.read_csv('covid_19_data.csv')

data
```

```python
# visualizing the missing values
import seaborn as sns
import matplotlib.pyplot as plt 

sns.heatmap(data.isnull(), cmap='Accent')
plt.xticks(rotation='vertical')
plt.show()
```

```python
data.info()
```

```python
# data.groupby('Region').value_counts()

data.groupby('Region').sum()
```

```python
# cases where the confirmed cases are more than 10

temp_data = data[~(data['Confirmed']<10)]

temp_data
```

```python
# region where max number of confirmed cases were recorded

temp_data = data.groupby("Region").sum()

temp_data[ temp_data['Confirmed'] == temp_data['Confirmed'].max() ]
```

```python
# region where min number of deaths were recorded

# I can use the .groupby() method to group the result based on Region
data[ data['Deaths'] == data['Deaths'].min() ].groupby("Region").value_counts()
```

```python
# cases from India till 29th April 2020

data[data['Region'] == 'India']
```

```python
# sorting data ascending order, based on confirmed cases 

data.groupby('Region')['Confirmed'].sum().sort_values(ascending=False)
```

```python
# sorting the whole data based on some column property

h_to_l_data = data.sort_values(by='Confirmed', ascending=True)

h_to_l_data
```

```python
# sorting in descending based on Recovered Cases 
recovered_cases = data.sort_values(by='Recovered', ascending=False)

recovered_cases
```
